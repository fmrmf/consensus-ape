```python
block = chain.blocks[-1]
block
```

```python
slots_per_epoch = 32
```

```python
# and for a query
# NOTE: start_block, stop_block are ** slot numbers **
df = chain.blocks.query('*', start_block=block.number - slots_per_epoch, stop_block=block.number)
df
```

```python
# expand out body ...
df['body'] = df['body'].apply(lambda x: x.dict())
```

```python
df[:10]
```

```python
# utility functions to unfold body arguments into separate columns
def unfold_args(key: str, args: dict) -> tp.Any:
    return args[key]

def unfold(df: pd.DataFrame):
    if df.empty:
        return
    for key in df.iloc[-1]['body'].keys():
        df[key] = df['body'].apply(lambda arg: unfold_args(key, arg))
```

```python
unfold(df)
```
